# Total teams year wise

```python
def year_wise(year):
    t1=IPL[IPL.season==year]['team1'].unique()
    t2=IPL[IPL.season==year]['team2'].unique()
    
    lst=[]
    for i in t1:
        lst.append(i)
    for i in t2:
        if i not in lst:
            lst.append(i)
    if lst==[]:
        return 'Not Found'
    else:
        return lst
        
```

```python
year_wise(2011)
```

```python
for i in IPL.season.unique():
    print(i,year_wise(i))
    print('*'*60)
    print()
```

# winning by runwise and season wise

```python
x=IPL.groupby(['season','team1']).win_by_runs.value_counts()
x.index
```

```python
def win_runs(year,team):
    return IPL[(IPL.season==year)&(IPL.winner==team)][['winner','win_by_runs','win_by_wickets']]
```

```python
win_runs(2016,'Kings XI Punjab')
```

# Winning by maximum runs 

```python
def max_runs(year,team):
    return win_runs(year,team)['win_by_runs'].max()
```

```python
max_runs(2008,'Royal Challengers Bangalore')
```

# Winning by maximum runs for all teams in 2008

```python
def team_win_by_runs(year):
    t=[]
    r=[]
    for i in IPL.winner.unique():
        x=max_runs(year,i)
        t.append(i)
        r.append(x)
    x=pd.DataFrame({'Team':t,'Win_by_maximum_runs':r})
    x.dropna(inplace=True)
    x=x.sort_values(by='Win_by_maximum_runs',ascending=False,ignore_index=True)
    return x
```

```python
team_win_by_runs(2008)
```

# Winning by Maximum runs in all seasons

```python
for year in range(2008, 2018):
    result = team_win_by_runs(year)
team_win_by_runs(year).max
print("In year",year,",win by Maximum run in all seasons")
print(team_win_by_runs(year))
```
